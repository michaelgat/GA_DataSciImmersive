#### Create a few new variables in your dataframe to represent interesting features of a job title.
- For example, create a feature that represents whether 'Senior' is in the title 
- or whether 'Manager' is in the title. 
- Then build a new Logistic Regression model with these features. Do they add any value? 


````python
##########################################################################
###                                                                    ###    
### Create dummy variables for various metrics that we are intersted   ###
### in exploring.                                                      ###
###                                                                    ### 
##########################################################################

# Create dummy variable for 'senior'
project4['senior'] = False
project4.senior = project4['title'].str.contains('Senior|senior|Sr|sr')
# Fix 'NaN' which results when the title is a 'Nan'
project4.senior.fillna(value = False, inplace = True)
project4.senior = project4.senior.astype(int)

# Create dummy variable for 'manager'
project4['manager'] = False
project4.manager = project4['title'].str.contains('Manager|manager|Mgr|mgr')
# Fix 'NaN' which results when the title is a 'Nan'
project4.manager.fillna(value = False, inplace = True)
project4.manager = project4.manager.astype(int)

# Create dummy variable for 'expert' and other upper level
project4['expert'] = False
project4.expert = project4['title'].str.contains('Expert|expert|Principal|principal')
# Fix 'NaN' which results when the title is a 'Nan'
project4.expert.fillna(value = False, inplace = True)
project4.expert = project4.expert.astype(int)

# Create dummy variable for 'machine learning'
project4['machlearn'] = False
project4.machlearn = project4['description'].str.contains('machine|Machine|Learning|learning')
# Fix 'NaN' which results when the desc is a 'Nan'
project4.machlearn.fillna(value = False, inplace = True)
project4.machlearn = project4.machlearn.astype(int)

# Create dummy variable for 'big data'
project4['bigdata'] = False
project4.bigdata = project4['description'].str.contains('Big data|big data|Big Data')
# Fix 'NaN' which results when the desc is a 'Nan'
project4.bigdata.fillna(value = False, inplace = True)
project4.bigdata = project4.bigdata.astype(int)

# Create dummy variable for PHD or MS
project4['PhDMS'] = False
project4.PhDMS = project4['description'].str.contains('Doctor| Dr| Doctorate| PhD| Masters| MS')
# Fix 'NaN' which results when the desc is a 'Nan'
project4.PhDMS.fillna(value = False, inplace = True)
project4.PhDMS = project4.PhDMS.astype(int)

project4.head(50)
````

````python
##########################################################################
###                                                                    ###    
### Create new dataframe to contain only those rows that have salary   ###
### information. Retain the previous dataset for use in other analysis ###
###                                                                    ### 
##########################################################################

project4_sal_only = project4.dropna(subset=['salary'])

project4_sal_only['clean_salary'] = project4_sal_only['salary'].replace('\D', '', regex=True).astype(int)

project4_sal_only.head()
````

````python
##########################################################################
###                                                                    ###    
### From Omar, split the salary where it represents a range            ###
### return the average of the range for use as our target value        ###
### this is returned in three columns that will need to be combined    ### 
###                                                                    ### 
##########################################################################

def spliti(i):
    """
    Takes salary range and splits it, then returns the average
    """
    i = str(i)
    leni=len(i)/2#works up to here
    j=i[0:leni]
    k=i[leni:len(i)]
    j=int(j)
    k=int(k)
    average = 0.5*(j+k)
    return average

a = project4_sal_only['clean_salary']

# Annual Salary Ranges
b = a[a > 1000000]
project4_sal_only['clean_salary1'] = b.apply(lambda x: spliti(x))

# Hourly Salary Ranges
d = a[a < 300]
project4_sal_only['clean_salary2'] = d.apply(lambda x: x*2000)

# Monthly Salary Ranges
f = a[(a < 25000) & (a > 10000)]
project4_sal_only['clean_salary3'] = f.apply(lambda x: x*12)

project4_sal_only.head()
````

````python
##########################################################################
###                                                                    ###    
### Now, combine and clean up all the columns as required.             ###
### Desired result is a single column for cleaned up salary.           ###
###                                                                    ### 
##########################################################################

def select_sal(row):
    """
    Return the value in the average salary column if it exists,
    which will only be the case if there was a salary range.
    Otherwise, return the cleaned up salary
    """   
    if row['clean_salary_all'] > 0:
        return row['clean_salary_all']
    else:
        return row['clean_salary']

# Take the highest value of each of the three columns created previously
# AT most, one of them will have a value, so this is trivial.
project4_sal_only['clean_salary_all'] = project4_sal_only[['clean_salary1','clean_salary2','clean_salary3']].max(axis=1)
project4_sal_only.head()
````

````python
# Use the average of a range if it exists. Otherwise use the salary as-is
project4_sal_only['salary_final'] = project4_sal_only.apply(select_sal, axis = 1)
project4_sal_only.head(50)
````

````python
# Drop all intermediate columns used for calculation (Whew!)
project4_sal_only.drop(['clean_salary', 'clean_salary1', 'clean_salary2', 'clean_salary3', 'clean_salary_all', 'salary'], inplace = True, axis = 1)

project4_sal_only.head(50)
````

````python
##########################################################################
###                                                                    ###    
### Summary statistics on salary that we will need going forward       ###
### These will be interesting at a national level, may also be useful  ###
### to calculate regionally if we get there in the analysis.           ###
###                                                                    ###
### Standard deviation will be important. To get target salary values  ###
### we might want a 25-50-25 split of salaries. The middle 40% lie     ###
### between -0.68 and 0.68 standard deviations of the mean.            ###
### This calcluation could be used on any subset of the data we        ###
### wish to use, including regional, by employer, etc.                 ###
###                                                                    ### 
##########################################################################

median = int(project4_sal_only['salary_final'].median())
mean = int(project4_sal_only['salary_final'].mean())
stdev = int(project4_sal_only['salary_final'].std())
low = mean - .68 * stdev
high = mean + .68 * stdev
print "Mean: " + str(mean)
print "Median: " + str(median)
print "StdDev: " + str(stdev)
print "High salary is above: " + str(high)
print "Low salary is below: " + str(low)

````

#### Rebuild this model with scikit-learn.
- You can either create the dummy features manually or use the `dmatrix` function from `patsy`
- Remember to scale the feature variables as well!


````python
# Dummy variables already created for "senior" "manager" "expert" "machine learning" and "big data"
# as well as other identified variants of those terms

from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix
import statsmodels.api as sm
import pylab as pl

cols_to_keep = ['senior', 'manager', 'expert', 'machlearn', 'bigdata', 'PhDMS', 'salary_final']
data = project4_sal_only[cols_to_keep]
data["intercept"] = 1.0
data['salary_high'] = np.where(data['salary_final'] >= mean, 1, 0)

train_cols = data.columns[0:6]

logit = sm.Logit(data["salary_high"], data[train_cols])

result = logit.fit()
result.summary2()
````

#### Use cross-validation in scikit-learn to evaluate the model above. 
- Evaluate the accuracy, AUC, precision and recall of the model. 
- Discuss the differences and explain when you want a high-recall or a high-precision model in this scenario.

````python
train_cols = ['machlearn', 'bigdata']

X = data[train_cols]
Y = data.salary_final
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33)

logreg1 = LogisticRegression(penalty='l1')
logreg1.fit(X_train, np.asarray(Y_train, dtype=np.int64))
Y_pred = logreg1.predict(X_test)
````

````python
X_train2, X_test2, Y_train2, Y_test2 = train_test_split(X, Y, test_size=0.33)

logreg2 = LogisticRegression(penalty='l2')
logreg2.fit(X_train2, np.asarray(Y_train2, dtype=np.int64))
Y_pred2 = logreg2.predict(X_test2)
````

#### Compare L1 and L2 regularization for this logistic regression model. What effect does this have on the coefficients learned?

````python
from sklearn.metrics import accuracy_score
print accuracy_score(Y_test.values.astype(int), Y_pred)
print accuracy_score(Y_test2.values.astype(int), Y_pred2)
````

````python
from sklearn.metrics import classification_report

print(classification_report(Y_test.values.astype(int), Y_pred))
````
